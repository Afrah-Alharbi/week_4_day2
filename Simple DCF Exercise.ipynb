```python
import numpy as np
import pandas as pd 
import pandas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
years=pandas.date_range(start='2022',end='2032',freq='Y')
years
```

```python
df = pd.DataFrame(index=years,columns=["Revenue", "EBITDA", "Taxes",
                           "Capital Expenditure", "Working Capital",
                           "Cash Flow", "Discount Factor",
                           "Present Value Free Cash Flow"], dtype='float')
df
```

```python
# Revenue Growth Factor: 18% of prior period Revenue
df.loc['2022',"Revenue"]=222*(1+0.18)
df
```

```python
for i in range(1,len(years)):
    df.iloc[i]["Revenue"]=df.iloc[i-1]["Revenue"]*(1+0.18)
df
```

```python
# EBITDA Margin: 45.0% of current Revenue
# Taxes: 21% of current EDBITDA
# Capital Expenditures: 3% of current Revenue
# Working Capital Increase: 6% of Revenue

df["EBITDA"]=df["Revenue"]*0.45
df["Taxes"]=-df["EBITDA"]*0.21
df["Capital Expenditure"]=-df["Revenue"]*0.03
df["Working Capital"]=-df["Revenue"]*0.06
df["Cash Flow"]=df.drop(columns="Revenue").sum(axis=1)
df 
```

```python
df["Present Value Free Cash Flow"] = df["Cash Flow"] * df["Discount Factor"]
df
```

```python
print(f"""Our projected Present Value of Projected Future Cash Flows are: {df["Present Value Free Cash Flow"].sum().round(0)}.
We are providing revenue guidance in 2032 to be: {round(df.iloc[-1]["Revenue"]*1.18)}.
With an estimated EBITDA around: {round(df.iloc[-1]["Revenue"]*1.18*0.45)}.

The terminal value at this time given a multiple of (3.70) and the above EBITDA for 2032, would equal: {round(df.iloc[-1]["Revenue"]*1.18*3.70)}.""")
```
